```python
import plotly
import plotly.graph_objs as go

plotly.offline.plot({
    "data": [go.Scatter(x=[1, 2, 3, 4], y=[4, 3, 2, 1])],
    "layout": go.Layout(title="hello world")
}, auto_open=True)
```

(読みやすかったらflaskでWebアプリ的なやつをするときにテンプレートとして使いたかった。)

# jupyterに埋め込む
 - graph_objsにグラフを構成する部品的なモノが用意されていて、そのリストをつくってplot(iplot)にぶち込めば良い
 - htmlで吐くときと埋め込むときで微妙に違うのが若干イケてない(plot⇒iplotだけだけど。)

```python
plotly.offline.init_notebook_mode(connected=True)

plotly.offline.iplot({
    "data": [go.Scatter(x=[1, 2, 3, 4], y=[4, 3, 2, 1])],
    "layout": go.Layout(title="hello world")
})
```

```python
import numpy as np

x = np.random.randn(500)
data = [go.Histogram(x=x)]

plotly.offline.iplot(data, filename='basic histogram')
```

# Matplotlib => plotly
 - とはいえmatplotlib歴8年やし、今更plotlyの語彙なんて覚えられないっすよーってなるわけじゃないですか。

```python
% matplotlib inline
from matplotlib import pyplot as plt
x=[1, 2, 3, 4]
y=[4, 3, 2, 1]
fig=plt.figure()
plt.plot(x,y,alpha=0.3)
plt.plot(x,y,'o')
plt.show()
```

```python
x=[1, 2, 3, 4]
y=[4, 3, 2, 1]
fig=plt.figure()
plt.plot(x,y,alpha=0.3)
plt.plot(x,y,'o')
plotly.offline.iplot_mpl(fig)
```

#  結論

これまでのコードに、
>import plotly
>plotly.offline.init_notebook_mode(connected=True)

して、

>plt.show()

の代わりに、

>plotly.offline.iplot_mpl(fig)

をこれからは使います。めでたしめでたし。
