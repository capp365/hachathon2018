```python
import json
f=open('./tweet.js','rt')
my_tweets=json.load(f)
```

```python
tw_texts=[i['full_text'] for i in my_tweets]
```

```python
import nagisa
tagged_texts=[nagisa.tagging(text) for text in tw_texts]
```

```python
import pickle
pickle.dump(tagged_texts,open('tagged_tweets','wb'))
```

```python
tagged_texts[0].words,tagged_texts[0].postags
```

```python
# そもそもPOS(Part-Of-Speech:品詞)タグ何があるのか
postags=list(set([j for i in tagged_texts for j in i.postags]))
print(postags)
```

```python
# postag出現回数
sorted([(k,[j for i in tagged_texts for j in i.postags].count(k)) for k in postags],key=lambda x:x[1])
```

```python
# 単語数(出現回数から重複排除)
w_count=sorted([(k,len(set([j[0] for i in tagged_texts for j in zip(i.words,i.postags) if j[1]==k]))) for k in postags],key=lambda x:x[1])
w_count
```

```python
for are in [(k[0],set([j[0] for i in tagged_texts for j in zip(i.words,i.postags) if j[1]==k[0]])) for k in w_count]:
    print(are[0])
    print(list(are[1])[:300])
```

```python
w_count2=sorted([(k,len(set([j[0] for i in tagged_texts for j in zip(i.words,i.postags) if j[1]==k]))\
                      ,len([j[0] for i in tagged_texts for j in zip(i.words,i.postags) if j[1]==k])) for k in postags],key=lambda x:x[1])
w_count2
```

```python
candidates=['接頭辞','代名詞','接尾辞','形状詞','形容詞','副詞','動詞','名詞']
```

```python
voc_cand=[[j for i in tagged_texts for j in zip(i.words,i.postags) if j[1] in k] for k in candidates]
```

```python
len(tagged_texts)
```

```python
countdic={i:{} for i in candidates}

for hoge,i in enumerate(tagged_texts):
    if hoge%10000==0:
        print(hoge)
    for j in zip(i.words,i.postags):
        if j[1] in candidates:
            if j[0] not in countdic[j[1]]:
                countdic[j[1]][j[0]]=1
            else:
                countdic[j[1]][j[0]]+=1
```

```python
vocab=[(j,i) for i in countdic for j in countdic[i] if countdic[i][j]>20]
len(vocab)
```
